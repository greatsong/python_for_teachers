CSV는 comma-separated values의 약자로 데이터 분석 분야에서는 가장 일반적인 파일 형식입니다.
이름에서 설명하는 것처럼 각 데이터 값들을 콤마 기호로 구분하는 특징을 갖고 있습니다.
아마 지금까지는 보신 적이 없을 수도 있지만 앞으로 남은 차시에서 많이 만나시게 될 겁니다.

그동안 많이 사용하셨던 엑셀 파일을 csv 파일로 바꾸는 것은 아주 간단한데요, 파일 내보내기에서 확장자를 csv라고 선택만 해주시면 됩니다.
사실 csv 파일을 만들거나 읽기 위해서는 메모장만 있어도 될 정도로 아주 쉽게 접근하실 수 있습니다.

그리고, 파이썬에서 이 csv 파일을 다룰 수 있도록 해주는 모듈이 csv 모듈입니다.

csv 모듈에는 크게 2개의 함수가 있는데요,
하나는 데이터를 읽어오는 reader 함수고,
다른 하나는 데이터를 기록하는 writer 함수입니다.

먼저 리더 함수에 대해 살펴보겠습니다.
리더 함수는 csv 파일에 저장된 데이터를 읽어오는 함수로 각 행의 데이터를 반복문에서 처리할 수 있는 형태로 돌려줍니다.
여러 옵션이 있지만 저희 연수에서는 복잡한 옵션을 사용하실 필요는 없기 때문에 기본적인 내용만 설명해드리겠습니다.

앞에서 보신 것처럼 열고자하는 csv 파일을 open 함수로 열고, 그것을 csv 모듈의 reader에서 읽어옵니다.

file = csv.reader(open('capital.csv'))

그러면 이것을 반복문으로 행별로 읽어올 수 있는데 이때 각 행에 저장된 데이터가 콤마 기호를 기준으로 나뉘어서 리스트로 저장이 됩니다.
따라서 다음과 같은 코드를 실행하면 한 행에 있는 데이터를 인덱스로 접근할 수 있게 됩니다.

for row in file : 
    print(row[0], row[-1])


다음은 롸이터 함수에 대해 알아보겠습니다.
롸이터 함수는 csv 파일에 데이터를 쓰기 위한 함수로 다음과 같이 사용됩니다.

note = csv.writer(open('note.csv', mode = 'w'))

그리고 실제로 csv 파일에 내용을 쓰기 위해서는 writerow 함수를 사용하는데, 다음과 같이 실행됩니다.

note.writerow(csv에 기록할 내용)

한 번 해결해보셨나요? 그러면 제가 해결한 방법을 설명해드리겠습니다.

지금 코드가 조금 복잡해보여서 저는 이 부분을 함수로 만들어보았습니다.

함수의 이름은 choice로 정했고, 함수의 역할은 정답인 도시 하나를 입력으로 받아서 4개의 보기를 출력해주는 함수입니다.

그리고 정답인 c가 포함된 리스트를 만듭니다.
그 다음엔 임의의 수도 이름을 리스트에 추가하는 것을 3번을 반복합니다.
이때, 임의의 수도 이름이 리스트 안에 있는 것이라면 중복된 것이기 때문에 while 반복문의 처음으로 돌아갑니다.
이렇게 3번 반복되면 정답이 포함된 4개의 수도 이름이 리스트로 만들어집니다.
그런데, 이 리스트에는 어떤 문제가 있을까요? 네, 항상 맨 처음 보기가 정답이 됩니다. 그래서 이 리스트를 랜덤 모듈의 셔플 함수로 섞어주었습니다.

이제 이 리스트를 순서대로 출력하면 보기가 보기 좋게 보이게 됩니다.

어떠셨나요? 지금까지 배웠던 내용들이 복합적으로 선택된 종합선물세트 같은 느낌의 프로젝트여서 많이 어려우셨을텐데요.
남은 시간에 배우실 내용에 오늘 배우신 내용이 큰 도움이 되실 겁니다.

```python
import csv
import random

# 정답인 도시 이름(c)를 입력받아서 4개의 보기를 출력해주는 함수
def choice(c) :
    #1. 정답인 c가 포함된 리스트 만들기
    L = [c]
    #2. 임의의 수도 이름 추가 : 3번 반복하기 
    cnt = 0
    while cnt < 3 :
        #3. 임의의 수도 이름 선택하기
        s = random.choice(list(capital.values()))
        #4. 선택한 수도 이름이 리스트에 있는 것일 경우 다시 선택하기
        if s in L :
            continue
        #5. 선택한 수도 이름 리스트에 추가하기
        L.append(s)
        cnt += 1
    #6. 4개의 수도 이름 임의로 섞기
    random.shuffle(L)
    
    #7. 4개의 보기 출력하기
    print('[보기]')
    for i in L :
        print(i)


file = csv.reader(open('capital.csv'))
note = csv.writer(open('note.csv', 'a'))

capital = {}

for row in file :
    capital[row[0]] = row[1]
    
n = random.choice(list(capital.keys()))
c = capital[n]

choice(c)

ans = input(str(n + '의 수도는?:'))
if ans == c :
    print('정답입니다!')
else :
    print(n+'의 수도는 '+c+'입니다. 오답 노트를 확인하세요!')
    #2. 오답노트에 틀린 내용 저장하기
    note.writerow([n,c])
```
